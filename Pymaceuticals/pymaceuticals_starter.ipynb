```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(study_results, mouse_metadata, how="outer", on="Mouse ID")

# Display the data table for preview
merge_df.head()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Here we are looking to see which mouse Ids have multiple same time points which would make them duplicates
# We then have it print out the mouse back to us
merge_df [merge_df.duplicated(subset= ["Mouse ID", "Timepoint"])]["Mouse ID"]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

merge_df [merge_df.duplicated(subset= ["Mouse ID", "Timepoint"])]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df[merge_df["Mouse ID"] != "g989"]
clean_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.


mean_vol = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_vol = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_vol = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_vol = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem_vol = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

statistics_df = pd.DataFrame()
statistics_df["Mean Tumor Volume"] = mean_vol
statistics_df["Median Tumor Volume"] = median_vol
statistics_df["Tumor Volume Variance"] = var_vol
statistics_df["Tumor Volume Std. Dev."] = std_vol
statistics_df["Tumor Volume Std. Err."] = sem_vol

statistics_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
clean_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var", "std", "sem"]})
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_cap = []
tumor_ram = []
tumor_inf = []
tumor_cef = []

# Calculate the IQR and quantitatively determine if there are any potential outliers for all the mice
quartiles = last_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Tumor Volume is: {lowerq}")
print(f"The upper quartile of Tumor Volume is: {upperq}")
print(f"The interquartile range of Tumor Volume is: {iqr}")
print(f"The the median of Tumor Volume is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
    
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# Created dataframe using .loc() of the drug regimine Capomulin using a boolean and keeping all the information from each of the columns
capo_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]

# Use groupby() to groupby mouse and mean() get the averages and create a new dataframe
capoavg = capo_df.groupby(['Mouse ID']).mean()

# Created scatter plot of the x axis (weight) and the y axis (tumor size) by calling from the capavg dataframe
plt.scatter(capoavg['Weight (g)'],capoavg['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)') #used to label the x axis
plt.ylabel('Average Tumor Volume (mm3)')# used to label the y axis

plt.show() #show the scatter plot graph
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Need to first calculate correlation
correlation = round(st.pearsonr(capoavg['Weight (g)'], capoavg['Tumor Volume (mm3)'])[0],2)
correlation
```

```python
# We now know that correlation is .84 which we know if it is close to 1 there is a high degree of positive correlation

# Need to now calculate the regression value using st.lineargress() function passing parameters from capo avg df
linear_regress=st.linregress(capoavg['Weight (g)'],capoavg['Tumor Volume (mm3)'])
linear_regress
```

```python
# Using the results of linear regress we can get the slope and the intercept
slope = 0.9544396890241049
intercept = 21.552160532685008
```

```python
# Now we will plot the regression line on the scatterplot

# calculate the linear line plot list of values
line = slope*(capoavg['Weight (g)'])+intercept

# create the scatter plot like above
plt.scatter(capoavg['Weight (g)'],capoavg['Tumor Volume (mm3)'])

# create the linear line on the graph
plt.plot(capoavg['Weight (g)'], line, 'r')

#label the x axis and y axis
plt.xlabel('Weight(g)')
plt.ylabel('Average Tumore Volume (mm3)')

# Print out the correlation statement and the graph with the linear line 
print(f'The correlation between mouse weight and the average tumor volume is {correlation}')
plt.show()
```
